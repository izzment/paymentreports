```python
df = pd.read_excel('20-25 Payments.xlsx')
org = pd.read_excel('Organizations.xlsx')
contr = pd.read_excel("Final.xlsx")
```

```python
contr = pd.read_excel("Final.xlsx")
contr = contr[["Beneficiary Name", "Date of Registration"]]
```

```python
contr = contr[["Beneficiary Name", "RC", "Registered Name"]]
```

```python
df = df.merge(contr, how="left", on="Beneficiary Name")
df.head()
```

```python
df["is_beneficiary_individual"] = (df["Type"] == "Individual").astype(int)
df.head()
```

```python
df["is_beneficiary_unknown"] = (df["Type"] == "Unknown").astype(int)
df.head()
```

```python
df["contractor_not_registered"] = np.where(
    df["Type"] == "Contractor", 
    df["RC"].notna().astype(int), 
    np.nan
)
df.head()
```

```python
df["contractor_not_registered"] = df["contractor_not_registered"].map({0: 1, 1: 0})
```

```python
df["has_description"] = (df["Description"].astype(str).str.strip().replace("nan", "").ne("")).astype(int)
df.head()
```

```python
df["has_description"] = 1 - df["has_description"]
```

```python
df[df["Type"]=="Contractor"]
```

```python
dup_key_cols = ["Payer Code", "Organization Name", "Beneficiary Name", "Amount", "Date", "Description"]
df["is_duplicate_payment"] = (df[dup_key_cols].duplicated(keep=False)).astype(int)
df.head()
```

```python
df[df[dup_key_cols].duplicated()]
```

```python
df["contractor_not_registered"] = df["contractor_not_registered"].fillna(0)
```

```python
# Weights
UNKNOWN_BENEFICIARY = 4
DUPLICATE_PAYMENT = 3
UNREGISTERED_CONTRACTOR = 2
NO_DESCRIPTION = 2
INDIVIDUAL_BENEFICIARY = 1

df["risk_score"] = (
     UNKNOWN_BENEFICIARY     * df["is_beneficiary_unknown"]
  + DUPLICATE_PAYMENT   * df["is_duplicate_payment"]
  + UNREGISTERED_CONTRACTOR * df["contractor_not_registered"]
  + NO_DESCRIPTION     * df["has_description"]
  + INDIVIDUAL_BENEFICIARY  * df["is_beneficiary_individual"]
).astype(int)
```

```python
df.head()
```

```python
df["risk_band"] = pd.cut(
    df["risk_score"],
    bins=[-1, 0, 1, 4, 999],
    labels=["LOW", "MILD", "CONCERNING", "HIGH"]
)
```

```python
df.to_csv("payments_enriched.csv", index=False)
```
