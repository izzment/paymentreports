```python
import pandas as pd

pd.set_option('display.float_format', '{:,.2f}'.format)
```

```python
df = pd.read_excel('2025 Payments.xlsx')
```

```python
dec = pd.read_excel('Unknown.xlsx')
```

```python
dec.head()
```

```python
dec['Date'] = pd.to_datetime(dec.Date)
```

```python
dec['Month'] = dec['Date'].dt.month
```

```python
mega = pd.read_excel('2020 to 2025 Gov Expenditure Dataset.xlsx')
```

```python
mega.head()
```

```python
mda_list = df[df.Type=='MDA']['Beneficiary Name'].unique()
con_list = df[df.Type=='Contractor']['Beneficiary Name'].unique()
ind_list = df[df.Type=='Individual']['Beneficiary Name'].unique()
```

```python
dec['Type'] = 'Unknown'

dec.loc[dec['Beneficiary Name'].isin(mda_list), 'Type'] = 'MDA'
dec.loc[dec['Beneficiary Name'].isin(con_list), 'Type'] = 'Contractor'
dec.loc[dec['Beneficiary Name'].isin(ind_list), 'Type'] = 'Individual'
```

```python
dec['Type'].value_counts()
```

```python
dec.to_excel('Unknown.xlsx', index=False)
```

```python
df.to_excel("2025 Payments.xlsx", index=False)
```

```python
orgs = pd.read_excel("Organizations.xlsx")
```

```python
org_list = orgs[['Organization Name','COFOG Code', "MDA", "Location", "Region", "Organization"]]
```

```python
orgs.head()
```

```python
df.head()
```

```python
df = df.merge(orgs[['Organization Name','COFOG Code', "MDA", "Location", "Region", "Organization"]], how='left', on='Organization Name')
```

```python
df = df.drop("Account", axis=1)
```

```python
df_list = [df, dec, mega]
```

```python
df = pd.concat(df_list, ignore_index=True)
```

```python
df.sort_values(by="Date", inplace=True)
```

```python
df[df['Year'] == 2025].head()
```
